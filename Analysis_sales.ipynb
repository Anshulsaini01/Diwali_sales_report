```python
Q1 = df_1['Amount'].quantile(0.25)
Q3 = df_1['Amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_clean = df_1[(df_1['Amount'] >= lower_bound) & (df_1['Amount'] <= upper_bound)]
```

```python
print("Original size:", len(df_1))
print("Cleaned size:", len(df_clean))
```

<!-- Which products generated the highest revenue during the Diwali sale period? -->

```python
# Group by product and sum the revenue
top_products = df_clean.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10)

# Reset index for plotting
top_products = top_products.reset_index()

# Set plot style
sns.set(style="whitegrid")

# Create barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Amount', y='Product_Category', data=top_products, palette='viridis')

# Add titles and labels
plt.title('Top 10 Revenue-Generating Products During Diwali Sale', fontsize=16)
plt.xlabel('Total Revenue (₹)', fontsize=12)
plt.ylabel('Product_Category', fontsize=12)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# - Which cities or regions contributed most to overall sales? 
```

```python
state_sales = df_clean.groupby('State')['Amount'].sum().sort_values(ascending=False)

# Set plot style
sns.set_style("whitegrid")

# Create pie chart
plt.figure(figsize=(8, 8))
colors = sns.color_palette('plasma', len(state_sales))
plt.pie(state_sales, labels=state_sales.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('State-wise Contribution to Diwali Sales', fontsize=14)
plt.axis('equal')  # Ensures pie is a perfect circle

# Show plot
plt.tight_layout()
plt.show()
```

```python
# What age group and gender spent the most during the Diwali sale?
```

```python
grouped_data = df_clean.groupby(['Age Group', 'Gender'])['Amount'].sum().reset_index()

# Set plot style
sns.set(style="whitegrid")

# Create grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_data, x='Age Group', y='Amount', hue='Gender', palette='coolwarm')

# Add titles and labels
plt.title('Total Spending by Age Group and Gender During Diwali Sale', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Total Amount Spent (₹)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Gender')

# Show plot
plt.tight_layout()
plt.show()
```
